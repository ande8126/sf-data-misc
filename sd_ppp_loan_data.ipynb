```python
import pandas as pd
```

```python
sf_ppp = ppp_loans[ppp_loans.City.str.contains('sioux falls', case = False)]
```

```python
sf_ppp.sort_values("LoanRange").head(36)
```

```python
sfloans = sf_ppp.groupby("LoanRange").LoanRange.count()
```

```python
sflenders = sf_ppp.groupby(["Lender", "LoanRange"]).size().reset_index(name="Activity")
```
